```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style='whitegrid')
%matplotlib inline
```

```python
sns.set(style='whitegrid', font_scale=1.1)
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
# Dataset path
path = r'C:\Users\Lenovo\OneDrive\Desktop\Employee-Attrition-Prediction model for clg\data\IBM-HR2.csv'

# CSV Load
df = pd.read_csv(path)

# Preview
df.head()
```

```python
print("\nshape:", df.shape)
print("\n Data Types:\n", df.dtypes)
```

```python
# Missing values check
print("\n‚ùó Missing Values:\n", df.isnull().sum())
```

```python
#  Unique values per column
print("\n Unique Values Per Column:\n", df.nunique())
```

```python
# Count of Attrition
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Attrition', palette='Set2')
plt.title("Attrition Distribution", fontsize=14, fontweight='bold')
plt.xlabel("Attrition", fontsize=12)
plt.ylabel("Count")
plt.show()
```

```python
# #Age Distribution by Attrition

plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Age', hue='Attrition', kde=True, palette='viridis', bins=30)
plt.title("Age Distribution vs Attrition", fontsize=14, fontweight='bold')
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
#  Boxplot - MonthlyIncome vs Attrition
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='Attrition', y='MonthlyIncome', palette='Set3')
plt.title("Monthly Income by Attrition", fontsize=14, fontweight='bold')
plt.xlabel("Attrition")
plt.ylabel("Monthly Income")
plt.show()
```

```python
#  Countplot - OverTime vs Attrition
plt.figure(figsize=(7, 5))
sns.countplot(data=df, x='OverTime', hue='Attrition', palette='viridis')
plt.title("OverTime vs Attrition", fontsize=14, fontweight='bold')
plt.xlabel("OverTime")
plt.ylabel("Count")
plt.legend(title='Attrition')
plt.show()
```

```python
#  Correlation Heatmap
plt.figure(figsize=(14, 10))
corr_matrix = df.corr(numeric_only=True)
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', linewidths=0.5)
plt.title("Feature Correlation Heatmap", fontsize=16, fontweight='bold')
plt.show()
```
